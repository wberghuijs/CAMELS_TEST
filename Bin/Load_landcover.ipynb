```python
# 3 lines to read the data
import pandas as pd
import matplotlib.pyplot as plt
data= pd.read_csv("example_catchment.csv")

# data contains: year', 'AFG', 'PFG', 'SHR', 'TRE', 'LTR', 'BGR', 'annualTemp','annualPrecip'

# make plot
fig, ax1 = plt.subplots()
color = 'tab:red'
ax1.set_ylabel('precipitation (mm/y)', color=color)
ax1.set_xlabel('year')
ax1.plot(data.year,data.annualPrecip, color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_xlim([1984,2022])
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis  
color1 = 'tab:blue'
color2 = 'tab:green'
color3 = 'tab:cyan'
color4 = 'tab:grey'
ax2.set_ylabel('landcover (%)', color=color1)  # we already handled the x-label with ax1
ax2.plot(data.year,data.TRE, color=color1, label='tree cover')
ax2.plot(data.year,data.BGR, color=color2, label='barren ground')
ax2.plot(data.year,data.SHR, color=color3, label='shrubs')
ax2.plot(data.year,data.AFG+data.PFG, color=color4, label='grassland')
ax2.tick_params(axis='y')
ax2.set_ylim([0, 100])
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.legend()
plt.show()
```
